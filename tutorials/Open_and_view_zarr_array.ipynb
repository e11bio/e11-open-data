```python
# pip install zarr s3fs matplotlib
import zarr, s3fs
import matplotlib.pyplot as plt

# Open the Zarr array from S3
fs = s3fs.S3FileSystem(anon=True)
store = s3fs.S3Map("e11bio-prism/ls/models/training_data/instance/crop_0.zarr", s3=fs)
# open zarr store.
f = zarr.open(store, mode="r")  # CZYX
# see all the arrays in the zarr store
print(f.tree())
```

```python
import matplotlib.pyplot as plt
# Show example crop from the 'raw' array
raw = f['raw'][0:3,100].T

# show data
plt.imshow(raw*5)
```
